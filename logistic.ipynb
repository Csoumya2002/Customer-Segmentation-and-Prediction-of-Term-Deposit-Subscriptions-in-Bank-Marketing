# Shapiro-Wilk normality test

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test
statistic, p_value = shapiro(df['age'])

# Print result
print(f"Shapiro-Wilk Test Statistic: {statistic:.4f}")
print(f"P-value: {p_value:.4f}")
# Interpretation
if p_value > 0.05:
    print("Data appears to be normally distributed (fail to reject H0).")
else:
    print("Data does not appear to be normally distributed (reject H0).")
```

```python
from scipy.stats import mannwhitneyu


# Split data into two groups
group1 = df[df['y'] == 'yes']['age']
group2 = df[df['y'] == 'no']['age']

# Perform Mann–Whitney U test
statistic, p_value = mannwhitneyu(group1, group2, alternative='two-sided')

# Print results
print(f"Mann–Whitney U Test Statistic: {statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value > 0.05:
    print("No significant difference between the two groups (fail to reject H0).")
else:
    print("Significant difference between the two groups (reject H0).")
```

```python
from scipy.stats import chi2_contingency
for col in categorical_vars:
    print(f"\n=== Chi-Square Test: {col} vs y ===")
    # Create contingency table
    table = pd.crosstab(df[col], df['y'])
    print("Contingency Table:")
    print(table)

    # Perform Chi-Square test
    chi2, p, dof, expected = chi2_contingency(table)
    print(f"\nChi2 Statistic: {chi2:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print(f"P-value: {p:.6e}")
```

### Cramer's V test

```python
def cramers_v(x, y):
    # Create the contingency table
    confusion_matrix = pd.crosstab(x, y)
    
    # Run chi-square test
    chi2, p, dof, expected = chi2_contingency(confusion_matrix)

    # Compute phi^2 and Cramer's V
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((k - 1)*(r - 1))/(n - 1))  # Bias correction
    r_corr = r - ((r - 1)**2)/(n - 1)
    k_corr = k - ((k - 1)**2)/(n - 1)
    
    # Return Cramér's V
    return np.sqrt(phi2_corr / min((k_corr - 1), (r_corr - 1)))

for var in categorical_vars:
    val = cramers_v(df[var], df['y'])
    print(f"Cramér's V ({var} vs y): {val:.4f}")
```

```python
numerical_vars = [
    'age', 'duration','campaign','pdays','previous','emp_var_rate','cons_price_idx','cons_conf_idx','nr_employed','euribor3m'
]
```

```python

import statsmodels.formula.api as smf
# Define target
df['y'] = df['y'].str.strip().str.lower().map({'no': 0, 'yes': 1})

# Combine all predictors
all_predictors = numerical_vars + categorical_vars

# Stepwise AIC function (same as before)
def stepwise_selection(data, response, predictors, direction='forward', verbose=True):
    remaining = set(predictors)
    selected = []
    current_score, best_new_score = float('inf'), float('inf')

    while remaining:
        scores_with_candidates = []
        for candidate in remaining:
            formula = f"{response} ~ {' + '.join(selected + [candidate])}"
            try:
                model = smf.logit(formula=formula, data=data).fit(disp=0)
                aic = model.aic
                scores_with_candidates.append((aic, candidate))
            except Exception as e:
                if verbose:
                    print(f"Skipping {candidate} due to error: {e}")
                continue

        if not scores_with_candidates:
            break

        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates[0]

        if current_score > best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
            if verbose:
                print(f"Adding {best_candidate:15} AIC = {best_new_score:.4f}")
        else:
            break

    # Final model
    final_formula = f"{response} ~ {' + '.join(selected)}"
    final_model = smf.logit(formula=final_formula, data=data).fit()
    return final_model

# Run stepwise selection
final_model = stepwise_selection(df, response='y', predictors=all_predictors)

# Show final model summary
print("\nFinal Logistic Regression Model:")
print(final_model.summary())
```

```python
aic_scores = []
for var in all_predictors:
    try:
        formula = f'y ~ {var}'
        model = smf.logit(formula=formula, data=df).fit(disp=0)
        aic_scores.append((var, model.aic))
    except Exception as e:
        print(f"Could not process {var}: {e}")

# Print AIC scores sorted from lowest (best) to highest
print("\nAIC Scores for Individual Predictors:")
for var, aic in sorted(aic_scores, key=lambda x: x[1]):
    print(f"{var:15} AIC = {aic:.4f}")
```

```python
import pandas as pd

results_table = pd.DataFrame({
    "Variable": final_model.params.index,
    "Estimate": final_model.params.values,
    "Std. Error": final_model.bse.values,
    "Z value": final_model.tvalues,
    "P value": final_model.pvalues
})

print(results_table)
```

```python
selected = []                            # variables included so far
remaining = all_predictors.copy()              # variables not yet added
current_aic = float('inf')               # start with infinite AIC

print("\n📊 Stepwise AIC Selection Process:\n")

while remaining:
    aic_candidates = []

    for candidate in remaining:
        formula = f"y ~ {' + '.join(selected + [candidate])}"
        try:
            model = smf.logit(formula=formula, data=df).fit(disp=0)
            aic = model.aic
            aic_candidates.append((aic, candidate))
        except Exception as e:
            print(f"Skipping {candidate}: {e}")

    if not aic_candidates:
        break

    # Sort by AIC (ascending)
    aic_candidates.sort()
    best_aic, best_candidate = aic_candidates[0]

    # Check if best candidate improves AIC
    if best_aic < current_aic:
        selected.append(best_candidate)
        remaining.remove(best_candidate)
        current_aic = best_aic
        print(f"✅ Added: {best_candidate:15} | AIC: {current_aic:.4f}")
    else:
        print("🔁 No further improvement. Stopping.")
        break

# Fit final best model
final_formula = f"y ~ {' + '.join(selected)}"
final_model = smf.logit(formula=final_formula, data=df).fit()

print("\n✅ Final Model Summary:")
print(final_model.summary())
```

```python
from sklearn.metrics import accuracy_score

# Step 1: Predict probabilities
pred_probs = final_model.predict(df)

# Step 2: Convert to binary class (threshold = 0.5)
pred_class = (pred_probs >= 0.5).astype(int)

# Step 3: Actual class
actual_class = df['y']

# Step 4: Compute accuracy
accuracy = accuracy_score(actual_class, pred_class)
print(f"\n✅ Model Accuracy: {accuracy:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Confusion matrix
cm = confusion_matrix(actual_class, pred_class)
print("\nConfusion Matrix:")
print(cm)

# Detailed classification report (Precision, Recall, F1)
print("\nClassification Report:")
print(classification_report(actual_class, pred_class))
```
